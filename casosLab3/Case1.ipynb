```python
import pandas
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv('caso1_0.1.csv')
```

```python
data
```

```python
time_gen = data['vectime'].loc[[26]].tolist()
buffer_gen = data['vecvalue'].loc[[26]].tolist()
time_queue = data['vectime'].loc[[27]].tolist()
buffer_queue = data['vecvalue'].loc[[27]].tolist()
time_sink = data['vectime'].loc[[28]].tolist()
buffer_sink = data['vecvalue'].loc[[28]].tolist()
x_time_delay = data['vectime'].loc[[29]].tolist()
y_time_delay = data['vecvalue'].loc[[29]].tolist()
time_packetDrops = data['vectime'].loc[[30]].tolist()
value_packetDrops = data['vecvalue'].loc[[30]].tolist()
```

```python
fig, ax = plt.subplots() #creo una figura y divido en pares para dibujar
plt.style.use("ggplot") #aplico un estilo de graficos
#plt.style.available -> para ver estilos posibles
ax.plot(time_sink,buffer_sink)
print(max(buffer_sink))
print(min(buffer_sink))
```

```python
fig, ax = plt.subplots() #creo una figura y divido en pares para dibujar
plt.style.use("ggplot") #aplico un estilo de graficos
#plt.style.available -> para ver estilos posibles
ax.plot(time_gen,buffer_gen)
print(max(buffer_gen))
print(min(buffer_gen))
```

```python
plt.plot(time_gen,buffer_gen, color='tab:blue', label="Tx")
plt.plot(time_queue,buffer_queue, color='tab:orange', label="Nx")
plt.plot(time_sink,buffer_sink, color='tab:green', label="Rx")
plt.title('Ocupacion de buffers en el sistema')
plt.xlabel("tiempo de simulacion")
plt.ylabel("Cantidad de paquetes en el buffer")
plt.legend()
plt.show
```

```python
plt.plot(time_sink,buffer_sink, color='tab:blue', label="Tx")
plt.plot(time_packetDrops, value_packetDrops, color='tab:green', label="PacketsDrops")
plt.title('Ocupacion de buffers en el sistema')
plt.xlabel("tiempo de simulacion")
plt.ylabel("Cantidad de paquetes en el buffer")
plt.legend()
plt.show
```
